```python
df = pd.read_csv('/kaggle/input/customer-clustering/segmentation data.csv')
df.head()
```

```python
df_cluster = df.drop(['ID','Sex','Marital status','Education','Occupation','Settlement size'],axis=1)
df_cluster.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_cluster['Age'],df_cluster['Income'])
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize = (10,7))
dendro = sch.dendrogram(sch.linkage(df_cluster,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=4,linkage='ward',affinity='euclidean')
label = cluster.fit_predict(df_cluster)
```

```python
label.shape
```

```python
plt.figure(figsize=(10,7))
plt.scatter(df_cluster['Age'],df_cluster['Income'],c=cluster.labels_,cmap='rainbow')
```
